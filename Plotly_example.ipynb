```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots
from datetime import datetime

import io
import pandas as pd
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
#print(uploaded.get(list(uploaded.keys())[0]))
uploaded_bytes = uploaded.get(list(uploaded.keys())[0])
df = pd.read_csv(io.StringIO(uploaded_bytes.decode('utf-8')), sep=';')
df['Date'] = pd.to_datetime(df['Date'])
print(type(df['Date'][0]))
df
```

```python
fig = go.Figure(data=[ go.Candlestick(x=df['Date'],
                                      open=df['LME open'], high=df['LME high'],
                                      low=df['LME low'], close=df['LME cash'])])
fig.show()
```

```python
idx = (df['Date'] > pd.Timestamp(2019,9,22)) & (df['Date'] < pd.Timestamp(2020,7,3))

df['Vol corrected'] = df["LME vol"][:]
df['Vol corrected'][idx] = df['Vol corrected'][idx]/10

fig = px.line(df, x='Date', y=['LME vol','Vol corrected'], markers=True)

fig.show()
```

```python
fig = plotly.subplots.make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    specs=[[{"type": "table"}],
           [{"type": "scatter"}],
           [{"type": "scatter"}]]
)

fig.add_trace(
    go.Table(
        header=dict(
            values=[k.replace(' ','<br>') for k in df.columns[:]],
            font=dict(size=10),
            align="left"
        ),
        cells=dict(
            values=[df[k].tolist() for k in df.columns[:]],
            align = "left")
    ),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(
        x=df["Date"],
        y=df["LME vol"],
        #mode="lines",
        name="Volume",
        line_color='#0000ff'
    ),
    row=2, col=1
)
fig.add_trace(
    go.Candlestick(x=df['Date'],
                   open=df['LME open'],
                   high=df['LME high'],
                   low=df['LME low'],
                   close=df['LME cash'],
                   name='Price'),
    row=3, col=1
)

fig.update_layout(
    height=1200,
    showlegend=True,
    title_text="<i><b>LME prices & volumes</b></i>",
)

fig.write_html("LME_prices_volumes.html")
fig.show()
```
